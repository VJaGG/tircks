### **1.PIL(RGB)**

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 8.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'
```

```python
image = Image.open('./imgs/image.jpg')
```

```python
print(type(image)) 
```

```python
print(image.size)  # (W, H)
```

```python
print(image.mode)
```

```python
image1 = image.convert('RGB')  # :rtype: :py:class:`~PIL.Image.Image`
                               # :returns: An :py:class:`~PIL.Image.Image` object
```

```python
image is image1
```

```python
type(image)
```

```python
type(image1)  # PIL.Image.Image-------->PIL.ImageFile.ImageFile--------->PIL.JpegImagePlugin.JpegImageFile
```

```python
print(image.getpixel((0, 0)))
```

```python
image.show()
```

```python
plt.figure()
# plt.axis('off')
plt.imshow(image)
```

```python
image = image.resize((int(504 / 2), int(744 / 2)), Image.NEAREST)
```

```python
plt.figure()
# plt.axis('off')
plt.imshow(image)
```

```python
print(image.size)  # w, h
```

```python
image_int = np.array(image)  # image=np.array(image)默认为uint8
image_float = np.array(image, dtype=np.float32)  #这样结果为小数
print(type(image_int))
print(image_int)
```

```python
print(image_float)
```

```python
print(image_int.shape)  # w和h换了，变成(h,w,c)了
                    # 注意ndarray中是 行row x 列col x 维度dim 所以行数是高，列数是宽
```

```python
print(image_float.shape)
```

```python
plt.figure()
plt.imshow(image_int)
# X : array-like or PIL image
#     The image data. Supported array shapes are:

#     - (M, N): an image with scalar data. The data is visualized
#       using a colormap.
#     - (M, N, 3): an image with RGB values (0-1 float or 0-255 int).
#     - (M, N, 4): an image with RGBA values (0-1 float or 0-255 int),
#       i.e. including transparency.

#     The first two dimensions (M, N) define the rows and columns of
#     the image.

#     Out-of-range RGB(A) values are clipped.
```

```python
plt.figure()
plt.imshow(image_float)
```

### **2.OpenCV(BGR)**

```python
import cv2
import numpy as np
```

```python
image = cv2.imread('./imgs/image.jpg')
```

```python
print(type(image))
```

```python
print(image.dtype)
```

```python
print(image.shape)  # (h, w, c)
```

```python
plt.figure()
plt.imshow(image)  # 根据颜色可以判别出OpenCV读入的数据为BGR
```

```python
image_BGR = Image.fromarray(image)  # 转换为PIL的格式
```

```python
print(type(image_BGR))
```

```python
print(image_BGR.mode)
```

```python
print(image_BGR.size)
```

```python
plt.figure()
plt.imshow(image_BGR)  # 这里的结果颜色发生了改变
```

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.figure()
plt.imshow(image)
```

 **PIL图像在转换为numpy.ndarray后，格式为(h, w, c)，像素的顺序为RGB  
 OpenCV在cv2.imread()后数据类型为numpy.ndarry，格式为(h, w, c)，像素的顺序为BGR**  
 

```python
print(image.shape)
```

```python
import torch
import torchvision.transforms as T
```

```python
transform = T.ToTensor()  # Converts a PIL Image or numpy.ndarray (H x W x C) in the range
                         # [0, 255] to a torch.FloatTensor of shape (C x H x W) in the range [0.0, 1.0]
                         # if the PIL Image belongs to one of the modes (L, LA, P, I, F, RGB, YCbCr, RGBA, CMYK, 1)
                         # or if the numpy.ndarray has dtype = np.uint8
```

```python
image_pil = Image.open('./imgs/image.jpg')
```

```python
# image_pil = image_pil.convert('RGB')  有没有效果都一样
```

```python
image_pil = transform(image_pil)
```

```python
# print(image_pil)
```

```python
image_cv = cv2.imread('imgs/image.jpg')
```

```python
image_cv_bgr = transform(image_cv)
```

```python
print(torch.sum(image_cv_bgr != image_pil))
```

```python
img_cv_rgb = cv2.cvtColor(image_cv, cv2.COLOR_BGR2RGB)
```

```python
img_cv_rgb = transform(img_cv_rgb)
```

```python
print(torch.sum(img_cv_rgb != image_cv_bgr))
```

```python
print(torch.sum(img_cv_rgb != image_pil))
```

```python
transform = T.ToPILImage()
```

```python
image_pil = transform(image_pil)
```

```python
plt.figure()
plt.imshow(image_pil)
```

```python
img_cv_rgb = transform(img_cv_rgb)
```

```python
plt.figure()
plt.imshow(image_cv_bgr)
```
